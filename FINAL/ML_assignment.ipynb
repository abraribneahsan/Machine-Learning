```python
import matplotlib.pyplot as plt
```

```python
semesters = list(range(1, 10))
cgpa = [3.57, 3.51, 3.56, 3.50, 3.47, 3.38, 3.28, 3.22, 3.15]
```

```python
plt.plot(semesters, cgpa, marker='o')
plt.title('CGPA Over Semesters')
plt.xlabel('Semester')
plt.ylabel('CGPA')
plt.grid(True)
plt.show()
```
